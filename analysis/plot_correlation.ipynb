```python
import os
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Constants
conversion_factor = 1
time_conversion_factor = 10

# Folder paths
local_folder = '/Users/scliu/Dropbox (Personal)/Academics/PhD_phase/Thomson_Lab/local to global pre-print/data/correlation_distance/local'
global_folder = '/Users/scliu/Dropbox (Personal)/Academics/PhD_phase/Thomson_Lab/local to global pre-print/data/correlation_distance/global'

# Read CSV files from the "local" and "global" folders
local_files = glob.glob(os.path.join(local_folder, "*.csv"))
global_files = glob.glob(os.path.join(global_folder, "*.csv"))

# Read and process local data
local_data = []
for file in local_files:
    data = pd.read_csv(file, header=None)
    data[1] = data[1] * conversion_factor
    data[0] = data[0] * time_conversion_factor / 60
    local_data.append(data)

local_mean = pd.concat(local_data).groupby(0).mean()
local_std = pd.concat(local_data).groupby(0).std()

# Read and process global data
global_data = []
for file in global_files:
    data = pd.read_csv(file, header=None)
    data[1] = data[1] * conversion_factor
    data[0] = data[0] * time_conversion_factor / 60
    global_data.append(data)

global_mean = pd.concat(global_data).groupby(0).mean()
global_std = pd.concat(global_data).groupby(0).std()


```

```python
# Plot the data
fig, ax = plt.subplots(figsize=(20, 5))

# Plot local data
ax.plot(local_mean.index, local_mean[1], label="Local")
ax.fill_between(local_mean.index, local_mean[1] - local_std[1], local_mean[1] + local_std[1], alpha=0.2)

# Plot global data
ax.plot(global_mean.index, global_mean[1], label="Global")
ax.fill_between(global_mean.index, global_mean[1] - global_std[1], global_mean[1] + global_std[1], alpha=0.2)
ax.set_xlim(0, 35)
ax.tick_params(axis='both', labelsize=14)

# Customize the plot
ax.set_xlabel("Time (min)", fontsize=16)
ax.set_ylabel("Correlation length (m)", fontsize=16)
ax.legend(fontsize=14)

# Show the plot
plt.show()
```
