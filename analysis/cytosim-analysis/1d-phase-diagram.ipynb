```python
# processing
import math
import scipy.stats as stats
import numpy as np
import pandas as pd

# visualization
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams.update({'font.size': 12})
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['font.sans-serif'] = "Arial"
```

### Functions

```python
def find_center(df):
    x = df['posX']
    y = df['posY']
    data = np.vstack((x, y))

    kde = stats.kde.gaussian_kde(data)
    gx, gy = np.mgrid[x.min():x.max():128j, y.min():y.max():128j]
    gxy = np.dstack((gx, gy)) # shape is (128, 128, 2)
    z = np.apply_along_axis(kde, 2, gxy)
    z = z.reshape(128, 128)

    center_ix, center_iy = np.unravel_index(z.argmax(), z.shape)
    center_x = center_ix / 128 * (x.max() - x.min()) + x.min()
    center_y = center_iy / 128 * (y.max() - y.min()) + y.min()
    return center_x, center_y
```

```python
def find_circle(x, xc, yc, r):
    y_upper = yc + np.sqrt(r**2 - (x - xc)**2)
    y_lower = yc - np.sqrt(r**2 - (x - xc)**2)
    return y_upper, y_lower
```

```python
def in_circle(row, xc_2, yc_2, R_2):
    x_coord = row['posX']
    if x_coord < (xc_2 - R_2) or x_coord > (xc_2 + R_2):
        return 0
    else:
        upper_bound, lower_bound = find_circle(x_coord, xc_2, yc_2, R_2)
        y_coord = row['posY']
        if y_coord > upper_bound or y_coord < lower_bound:
            return 0
        else:
            return 1
```

```python
def get_alpha(dirX, dirY, L):
    alpha = L/2 / math.sqrt(dirX**2 + dirY**2)
    return alpha
```

```python
def get_end_coords(posX, posY, dirX, dirY, alpha):
    startX = posX - alpha * dirX
    startY = posY - alpha * dirY
    return startX, startY
```

```python
def get_nodes(posX, posY, dirX, dirY, L):
    alpha = get_alpha(dirX, dirY, L)
    startX, startY = get_end_coords(posX, posY, dirX, dirY, alpha)
    segmentX = 2 * alpha * dirX / L
    segmentY = 2 * alpha * dirY / L

    nodes = [(startX, startY)]
    for i in range(1, L+1):
        nodeX = startX + i * segmentX
        nodeY = startY + i * segmentY
        nodes.append((nodeX, nodeY))

    return nodes
```

```python
def cartesian_dist(n1, n2):
    return math.sqrt((n1[0] - n2[0])**2 + (n1[1] - n2[1])**2)
```

```python
def get_min_dist(nodes1, nodes2):
    min_dist = 10**6
    for n1 in nodes1:
        for n2 in nodes2:
            d = cartesian_dist(n1, n2)
            if d < min_dist:
                min_dist = d
    return min_dist
```

```python
def get_mean_min_dist(df):
    microtubules = []
    for index, row in df.iterrows():
        posX = row['posX']
        posY = row['posY']
        dirX = row['dirX']
        dirY = row['dirY']
        L = int(row['length'])
        nodes = get_nodes(posX, posY, dirX, dirY, L)
        microtubules.append(nodes)
    
    list_min_dist = []
    for i, nodes1 in enumerate(microtubules):
        for j, nodes2 in enumerate(microtubules[(i+1):]):
            min_dist = get_min_dist(nodes1, nodes2)
            list_min_dist.append(min_dist)
    
    return np.mean(list_min_dist)
```

### Processing all the simulations

```python
df_list = []
for i in range(0, 10):
    for j in range(0, 20):
        num_zeros = 4 - len(str(j))
        fdir = '/Users/rosalindpan/thomsonlab/simulation/ms-simulations/1d-phase/all-runs/series{}/run0000-{}/'.format(int(i), num_zeros * '0' + str(j))

        rv = {}

        with open(fdir + 'config.cym') as fp:
            lines = fp.readlines()[0]
            lines_split = lines.strip().split(' ')
            shape = lines_split[2]
            scale = lines_split[-1]
            rv['shape'] = shape.strip(',')
            rv['scale'] = scale
        
        fpath = fdir + 'fiber_position.txt'
        df = pd.read_table(fpath,
                           skiprows=5,
                           delim_whitespace=True,
                           names=['class', 'identity', 'length', 'posX', 'posY',
                                  'dirX', 'dirY', 'endToEnd', 'cosinus', 'organizer'])
        df = df.dropna(how='any')

        center_x, center_y = find_center(df)
        col = df.apply(in_circle, args=(center_x, center_y, np.median(df['length'])), axis=1)
        df = df.assign(in_circle=col.values)

        rv['aster_fraction'] = sum(df['in_circle']) / len(df)
        rv['mean_min_dist'] = get_mean_min_dist(df)
        df_list.append(rv)
```

```python
df = pd.DataFrame(df_list).sort_values(by='scale')
df['scale'] = pd.to_numeric(df['scale'])
df_agg = df.groupby(['scale'], as_index=False).agg({'aster_fraction':['mean','std'], 'mean_min_dist':['mean','std']})
```

```python
df_agg.columns = ['scale', 'aster_fraction_mean', 'aster_fraction_std', 'mean_min_dist_mean', 'mean_min_dist_std']
```

```python
fig, axes = plt.subplots(2, 1, figsize=(9,6), sharex=True)
ax1, ax2 = axes
fig.patch.set_facecolor('white')

scale_values = df_agg['scale'].values
frac_mean = df_agg['aster_fraction_mean'].values
frac_std = df_agg['aster_fraction_std'].values
dist_mean = df_agg['mean_min_dist_mean'].values
dist_std = df_agg['mean_min_dist_std'].values

scale = np.arange(0.25, 5.1, 0.25)
ax1.scatter(scale_values, frac_mean, color='rebeccapurple')
ax1.fill_between(scale_values, frac_mean - frac_std, frac_mean + frac_std,
                 color='rebeccapurple', alpha=0.2)
ax1.set_xticks(np.arange(0, 5.1, 0.5))
ax1.set_yticks(np.arange(0, 0.81, 0.2))
ax1.set_yticklabels([0, 0.2, 0.4, 0.6, 0.8], fontsize=15)
ax1.set_ylabel('Fraction of MTs\nin the largest aster', fontsize=20)

ax2.scatter(scale_values, dist_mean, color='forestgreen')
ax2.fill_between(scale_values, dist_mean - dist_std, dist_mean + dist_std,
                 color='forestgreen', alpha=0.2)
ax2.set_xlabel('Scale of gamma distribution', fontsize=20)
ax2.set_ylabel('Mean minimum\ndistance', fontsize=20)
ax2.set_yticks(np.arange(0, 15.01, 2.5))
ax2.set_yticklabels(np.arange(0, 15.01, 2.5), fontsize=15)
ax2.set_xticklabels(np.arange(0, 5.1, 0.5), fontsize=15)

plt.tight_layout()
plt.savefig('../../figures/figure3/fig3A.png', dpi=300)
plt.show()
```
