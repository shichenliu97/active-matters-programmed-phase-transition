```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('../../experiments/correlation_data/5MT2ul.csv', header=None, names=['frame', 'dist'])
df2 = pd.read_csv('../../experiments/correlation_data/130MT45ul.csv', header=None, names=['frame', 'dist'])
df3 = pd.read_csv('../../experiments/correlation_data/205MT25ul.csv', header=None, names=['frame', 'dist'])
```

```python
fig, ax = plt.subplots()

ax.plot(df1['frame'], df1['dist'])
ax.plot(df2['frame'], df2['dist'])
ax.plot(df3['frame'], df3['dist'])

ax.legend([''])

plt.show()
```
